First of all we need to import Pandas library.


```python
import pandas as pd
```

DataFrame:
It is the table. It is array of individual entries.


```python
pd.DataFrame({"a":[1,2,3],"b":[4,5,6]})
```

```python
pd.DataFrame({"a":["I am Ilyan","I am learning pandas."]})
```

This is used with dictionaries.
keys = colomns
values = rows

We can change the index from starting from 0 to any other index.


```python
pd.DataFrame({"A":[1,2,3],"B":[4,5,6]},index=["1","2","3"])
```

Series:
It is just a list of values.
It has only 1 colomn.


```python
pd.Series([1,2,3],index=["a","b","c"],name="My_Series")
```
