```python
import numpy as np
```

```python
#scaler arthemetic = it is the operation between a single number and an array.
array1=np.array([1,2,3,4])
print(array1+1)
print(array1-1)
print(array1*2)
print(array1/2)
print(array1**2)
```

```python
#vectorized math functions = it is the operation between two arrays of the same size.
array2=np.array([1,2,3,4.9])
print(np.sqrt(array2))
print(np.round(array2))
print(np.pi)
```

```python
radii=np.array([1,2,3,4])
print(np.pi*radii**2)
```

```python
# element wise operations = it is the operation between two arrays of the same size.
array3=np.array([1,2,3])
array4=np.array([4,5,6])

print(array3+array4)
print(array3-array4)
print(array3*array4)
print(array3/array4)
```

```python
# comparasion operators
scores=np.array([90,80,70,60,100])
# print(scores ==100)
print(scores >=70)
scores[scores <70]=0     #change the values that are less than 70 to 0
print(scores)
```
